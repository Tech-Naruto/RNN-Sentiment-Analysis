```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# >>> Sentences <<<

sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'not understanding of words'
]
```

```python
# >>> Vocabulary Size <<<

vocab_size = 10000
```

```python
#  >>> One Hot Encoding <<<

one_hot_repr = [one_hot(text, vocab_size) for text in sent]
one_hot_repr
```

```python
# >> Word Embedding Representation <<<

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
# >>> Padding <<<

max_sent_length = 8
padded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=max_sent_length)
padded_docs
```

```python
# >>> Feature representation <<<

dim = 10

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=dim))
model.compile('adam', 'mse')
```

```python
model.predict(padded_docs)
```

```python
model.summary()
```
